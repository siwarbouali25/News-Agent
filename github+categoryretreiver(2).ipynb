```python
# === Cloudflare Tunnel version: Flask webhook + FAISS builder ===
!pip -q install flask langchain-community sentence-transformers faiss-cpu requests

import os, io, json, base64, time, sys, requests, pandas as pd, threading, pathlib, subprocess, re
from flask import Flask, request, jsonify, abort
from langchain.schema import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from google.colab import drive

# ------------------ Drive & paths ------------------
drive.mount('/content/drive')
DRIVE_DIR   = "/content/drive/MyDrive/newsbot_shared"
INDEX_DIR   = f"{DRIVE_DIR}/faiss_news_index"
CSV_PATH    = f"{DRIVE_DIR}/Articles.csv"
STATUS_PATH = f"{DRIVE_DIR}/embed_status.json"
LOG_PATH    = f"{DRIVE_DIR}/webhook.log"
pathlib.Path(DRIVE_DIR).mkdir(parents=True, exist_ok=True)

# ------------------ Repo / model -------------------
OWNER, REPO, PATH, BRANCH = "miriambenouaghrem", "test_rep", "Articles.csv", "main"
EMBED_MODEL = "sentence-transformers/all-MiniLM-L6-v2"

# Optional simple shared secret (set same in your GitHub Action as WEBHOOK_TOKEN)
WEBHOOK_TOKEN = os.environ.get("WEBHOOK_TOKEN", "changeme")

def log(msg: str):
    ts = time.strftime("%Y-%m-%d %H:%M:%S UTC", time.gmtime())
    line = f"[{ts}] {msg}"
    print(line); sys.stdout.flush()
    with open(LOG_PATH, "a") as f:
        f.write(line + "\n")

def fetch_csv():
    url = f"https://api.github.com/repos/{OWNER}/{REPO}/contents/{PATH}?ref={BRANCH}"
    r = requests.get(url, headers={"Accept": "application/vnd.github+json"}, timeout=30)
    r.raise_for_status()
    js = r.json()
    if js.get("encoding") == "base64" and js.get("content"):
        text = base64.b64decode(js["content"]).decode("utf-8", errors="replace")
    else:
        dl = requests.get(js["download_url"], timeout=30); dl.raise_for_status()
        text = dl.text
    return js["sha"], text

def build_embeddings(text: str, sha: str):
    # robust CSV read
    try:
        df = pd.read_csv(io.StringIO(text))
    except Exception:
        df = pd.read_csv(io.StringIO(text), engine="python", on_bad_lines="skip")
    df = df.fillna("")

    # ensure required columns
    for c in ["title","content"]:
        if c not in df.columns:
            raise ValueError(f"CSV missing required column: {c}")

    # mirror CSV to Drive
    df.to_csv(CSV_PATH, index=False)

    # build FAISS
    emb = HuggingFaceEmbeddings(model_name=EMBED_MODEL)
    docs = [
        Document(
            page_content=str(r["title"]) + "\n\n" + str(r["content"]),
            metadata={k: str(r[k]) for k in df.columns if k != "content"},
        )
        for _, r in df.iterrows()
    ]
    pathlib.Path(INDEX_DIR).mkdir(parents=True, exist_ok=True)
    vs = FAISS.from_documents(docs, emb)
    vs.save_local(INDEX_DIR)

    # write status
    pathlib.Path(STATUS_PATH).write_text(json.dumps({
        "sha": sha, "rows": int(len(df)), "last_embed_ts": time.time()
    }))

    log(f"âœ… FAISS rebuilt â€¢ rows={len(df)} â€¢ sha={sha[:7]}")

# ------------------ Flask app ----------------------
app = Flask(__name__)

@app.get("/healthz")
def healthz():
    return jsonify(ok=True, time=time.time())

@app.post("/refresh")
def refresh():
    # simple auth header
    if request.headers.get("X-Webhook-Token") != WEBHOOK_TOKEN:
        abort(401)
    log("ğŸš€ Webhook /refresh received")
    try:
        sha, text = fetch_csv()
        build_embeddings(text, sha)
        return jsonify(status="ok", sha=sha[:7])
    except Exception as e:
        log(f"âŒ Error: {e}")
        return jsonify(status="error", detail=str(e)), 500

# Start Flask on a clean port (8081)
PORT = 8081
def _run_flask():
    app.run(port=PORT)

threading.Thread(target=_run_flask, daemon=True).start()

# ------------------ Cloudflare Tunnel ------------------
# No account needed. This prints a stable (per-session) https://*.trycloudflare.com URL.
!wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O cloudflared
!chmod +x cloudflared

# kill any old cloudflared in this runtime (best-effort)
try:
    subprocess.run(["pkill","-f","cloudflared"], check=False)
except Exception:
    pass

proc = subprocess.Popen(
    ["./cloudflared","tunnel","--url", f"http://localhost:{PORT}", "--no-autoupdate"],
    stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True
)

# read stdout until the public URL appears
public_url = None
for line in proc.stdout:
    m = re.search(r'https://[-\w]+\.trycloudflare\.com', line)
    if m:
        public_url = m.group(0)
        break

if not public_url:
    raise RuntimeError("Could not detect Cloudflare public URL from cloudflared output.")

print("HEALTH :", public_url + "/healthz")
print("REFRESH:", public_url + "/refresh")
log(f"ğŸŒ Webhook URL (Cloudflare): {public_url}/refresh")
```

```python
%%writefile app.py
import os, re, io, json, time, pathlib, pandas as pd
import streamlit as st
from urllib.parse import urlparse
from typing import TypedDict, List, Dict, Any
import torch
from datetime import datetime

from langgraph.graph import StateGraph, END
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig

# ================================
# DRIVE LOCATIONS (must match producer)
# ================================
DRIVE_DIR   = "/content/drive/MyDrive/newsbot_shared"
INDEX_DIR   = f"{DRIVE_DIR}/faiss_news_index"
STATUS_FILE = f"{DRIVE_DIR}/embed_status.json"
CSV_PATH    = f"{DRIVE_DIR}/Articles.csv"       # optional preview only

EMBED_MODEL = "sentence-transformers/all-MiniLM-L6-v2"
MODEL_ID    = "meta-llama/Meta-Llama-3-8B-Instruct"   # swap to an ungated model if you don't use a HF token
DATE_COL    = "published_date"

# Try to mount Drive when running inside Colab
try:
    import google.colab  # type: ignore
    from google.colab import drive
    if not os.path.exists("/content/drive/MyDrive"):
        drive.mount("/content/drive")
except Exception:
    pass

# ================================
# PAGE & STYLE
# ================================
st.set_page_config(page_title="News Agent", layout="wide")
st.markdown("""
<style>
:root{ --bg:#0f1116; --bg-soft:#151822; --text:#e6e9ef; --muted:#8b91a1; --brand:#35c2c1; --border:#2a2f3b; }
html, body, [data-testid="stAppViewContainer"]{ background:var(--bg); color:var(--text); }
[data-testid="stSidebar"]{ background:var(--bg-soft)!important; }
h1,h2,h3{ color:var(--text) }
a, a:visited{ color:var(--brand); text-decoration:none; }
.card{ border:1px solid var(--border); background:linear-gradient(180deg,#141826,#121623);
border-radius:18px; padding:16px; height:100%; box-shadow:0 0 0 1px rgba(53,194,193,.08), 0 8px 24px rgba(0,0,0,.35); }
.card .title{ font-weight:700;font-size:1.05rem;line-height:1.3;margin:0 0 6px;color:var(--text); }
.card .meta{ color:var(--muted);font-size:.9rem;margin:0 0 10px; }
.card a.open{ display:inline-block;color:var(--brand);font-weight:600; }
.card a.open:hover{ text-decoration:underline; }
.section{ margin: 8px 0 22px; font-weight:800;font-size:1.25rem;letter-spacing:.2px; }
.section .tag{ color:var(--muted);font-weight:600;font-size:1rem;margin-left:.25rem; }
hr{border-color:var(--border)!important}
.badge{font-size:.85rem;color:#cfd3dc}
</style>
""", unsafe_allow_html=True)

st.title("ğŸ“° News AI Agent")

# ================================
# STATUS / AUTO-RELOAD KEY
# ================================
def _status() -> dict:
    p = pathlib.Path(STATUS_FILE)
    if not p.exists():
        return {}
    try:
        return json.loads(p.read_text())
    except Exception:
        return {}

def _dir_mtime(path: str) -> float:
    """Return latest mtime of a directory tree (0 if missing)."""
    try:
        base = pathlib.Path(path)
        if not base.exists():
            return 0.0
        mt = 0.0
        for root, _, files in os.walk(base):
            for f in files:
                try:
                    mt = max(mt, os.path.getmtime(os.path.join(root, f)))
                except Exception:
                    pass
        return mt
    except Exception:
        return 0.0

def _status_key() -> str:
    s = _status()
    stat_part  = str(s.get("last_embed_ts") or s.get("ts") or s.get("sha") or "none")
    faiss_part = f"{_dir_mtime(INDEX_DIR):.0f}"
    csv_part   = f"{os.path.getmtime(CSV_PATH):.0f}" if os.path.exists(CSV_PATH) else "0"
    return "|".join([stat_part, faiss_part, csv_part])

def pretty_date(ds):
    if not ds or ds in ["nan", "NaT"]:
        return "Unknown date"
    try:
        dt = datetime.fromisoformat(str(ds).replace("Z","+00:00"))
        return dt.strftime("%Y-%m-%d %H:%M UTC")
    except Exception:
        return str(ds)

# ================================
# RETRIEVER (LOAD ONLY, NEVER REBUILD HERE)
# cache key = status key, so it reloads when producer updates Drive
# ================================
@st.cache_resource
def load_retriever(_key: str):
    if not os.path.isdir(INDEX_DIR):
        raise FileNotFoundError(f"FAISS index directory not found: {INDEX_DIR}")
    emb = HuggingFaceEmbeddings(model_name=EMBED_MODEL)
    vs  = FAISS.load_local(INDEX_DIR, emb, allow_dangerous_deserialization=True)
    return vs.as_retriever(search_kwargs={"k": 10})

# CSV preview keyed to status key
@st.cache_data
def _load_csv(_key: str) -> pd.DataFrame:
    if not os.path.exists(CSV_PATH):
        raise FileNotFoundError(CSV_PATH)
    return pd.read_csv(CSV_PATH)

# ---- Detect new embeddings and auto-clear caches/results
current_key = _status_key()
prev_key = st.session_state.get("emb_key")

if prev_key != current_key:
    # new embeddings â†’ drop old results and refresh only retriever/CSV caches
    st.session_state.pop("articles_by_cat", None)
    st.session_state["emb_key"] = current_key

    # ONLY clear caches tied to data; keep LLM cached
    try:
        load_retriever.clear()   # <- specific cache
        _load_csv.clear()        # <- specific cache
    except Exception:
        pass
    st.rerun()

# after possible rerun, we have a stable key in session
key = st.session_state.get("emb_key", current_key)

# Helper that always returns a retriever bound to the *current* key
def _get_retriever():
    return load_retriever(key)

# ================================
# LLM
# ================================
@st.cache_resource
def load_llm():
    bnb = BitsAndBytesConfig(
        load_in_4bit=True, bnb_4bit_quant_type="nf4",
        bnb_4bit_use_double_quant=True, bnb_4bit_compute_dtype=torch.bfloat16,
    )
    tok = AutoTokenizer.from_pretrained(MODEL_ID)
    mdl = AutoModelForCausalLM.from_pretrained(
        MODEL_ID, device_map="auto", quantization_config=bnb, torch_dtype=torch.bfloat16
    )
    return pipeline("text-generation", model=mdl, tokenizer=tok,
                    max_new_tokens=200, temperature=0.3, do_sample=False)

def get_llm():
    # Lazy: only construct when needed
    if "llm_pipe" not in st.session_state:
        st.session_state["llm_pipe"] = load_llm()
    return st.session_state["llm_pipe"]


# Load resources
try:
    retriever = _get_retriever()
except Exception as e:
    st.error("Could not load FAISS retriever. Make sure the producer has written the index to Drive.")
    st.exception(e)
    st.stop()

#llm_pipe  = load_llm()

# Top status bar
s = _status()
rows = s.get("rows", "â€”")
sha  = (s.get("sha") or "")[:7]
ts   = s.get("last_embed_ts") or s.get("ts")
updated = datetime.utcfromtimestamp(ts).strftime("%Y-%m-%d %H:%M UTC") if ts else "â€”"
st.caption(f"**Data status** Â· rows: {rows} Â· sha: `{sha}` Â· updated: {updated} Â· key: `{key}`")

# Optional CSV preview (keyed, so it refreshes when data changes)
with st.expander("Preview latest Articles.csv"):
    try:
        st.dataframe(_load_csv(key).head(20), use_container_width=True)
    except Exception as e:
        st.info("CSV preview not available yet.")
        st.exception(e)

# ================================
# AGENT LOGIC
# ================================
class State(TypedDict):
    query: str
    k: int
    mode: str
    results: List[Dict[str, Any]] | List[Document]
    answer: str

def format_hits(hits: List[Document], k: int) -> List[Dict[str, Any]]:
    out, seen = [], set()
    for d in hits:
        meta = d.metadata or {}
        url  = meta.get("url", "")
        domain = urlparse(url).netloc
        if domain in seen:   # one per domain
            continue
        seen.add(domain)
        out.append({
            "title": meta.get("title", ""),
            "source": meta.get("source", ""),
            "published_date": meta.get("published_date", ""),
            "url": url
        })
        if len(out) >= k:
            break
    return out

def route(state: State) -> str:
    q = (state["query"] or "").lower().strip()
    if q.endswith("?") or re.match(r"^(what|who|why|how|when|where)\\b", q):
        state["mode"] = "qa"; return "qa"
    state["mode"] = "list"; return "list"

def router_node(state: State) -> State:
    route(state);
    return state

def node_retrieve_list(state: State) -> State:
    docs = _get_retriever().get_relevant_documents(state["query"])
    state["results"] = format_hits(docs, state["k"]);
    return state

def node_retrieve_qa(state: State) -> State:
    docs = _get_retriever().get_relevant_documents(state["query"])
    state["results"] = docs[:6];
    return state

def node_llm_answer(state: State) -> State:
    docs: List[Document] = state["results"]
    if not docs:
        state["answer"] = "I couldn't find any relevant articles to answer that."
        return state

    # use the cached LLM instance (never reloaded)
    llm = get_llm()
    if llm is None:
        state["answer"] = "LLM is not loaded yet. Please wait a moment and retry."
        return state

    ctx = []
    for d in docs:
        m = d.metadata or {}
        title = m.get("title", "")
        ctx.append(f"[{title}] {d.page_content[:400]}")

    prompt = (
        "You are a concise journalist assistant. Use only the CONTEXT below. "
        "Cite titles in parentheses when relevant.\n\nCONTEXT:\n"
        + "\n\n".join(ctx)
        + f"\n\nQUESTION: {state['query']}\n\nANSWER:"
    )

    out = llm(prompt)[0]["generated_text"]
    state["answer"] = out.split("ANSWER:", 1)[-1].strip()
    return state


g = StateGraph(State)
g.add_node("router", router_node); g.set_entry_point("router")
g.add_conditional_edges("router", route, {"list": "retrieve_list", "qa": "retrieve_qa"})
g.add_node("retrieve_list", node_retrieve_list)
g.add_node("retrieve_qa", node_retrieve_qa)
g.add_node("llm_answer", node_llm_answer)
g.add_edge("retrieve_list", END); g.add_edge("retrieve_qa", "llm_answer"); g.add_edge("llm_answer", END)
graph = g.compile()

# ================================
# UI
# ================================
all_categories = ["Technology", "World", "Politics", "Science", "Health", "Sports", "Culture"]
selected = st.sidebar.multiselect("Choose categories", all_categories, default=["Technology","Politics"])
k = st.sidebar.slider("Articles per category", 1, 10, 5)

# Auto-refresh every 30s so key changes trigger re-run (helps live dashboards)
try:
    from streamlit_autorefresh import st_autorefresh
    st_autorefresh(interval=30_000, key="auto")
    st.caption("Auto-refresh every 30s (reloads when producer updates).")
except Exception:
    pass

if st.sidebar.button("Get Top Articles"):
    # Use latest retriever at click-time
    retr = _get_retriever()
    by_cat = {}
    for cat in selected:
        docs = retr.get_relevant_documents(cat)
        by_cat[cat] = format_hits(docs, k)
    st.session_state["articles_by_cat"] = by_cat

if "articles_by_cat" in st.session_state:
    for cat, items in st.session_state["articles_by_cat"].items():
        st.markdown(f"<div class='section'>{cat} â€” {len(items)} most recent <span class='tag'>(diverse sources)</span></div>", unsafe_allow_html=True)
        if not items:
            st.caption("No articles found.")
            continue
        cols = st.columns(3)
        for i, a in enumerate(items):
            with cols[i % 3]:
                title = a.get("title") or "(Untitled)"
                url   = a.get("url") or ""
                src   = a.get("source") or "Unknown source"
                dt    = a.get(DATE_COL)
                st.markdown(f"""
                  <div class="card">
                    <div class="title">{title}</div>
                    <div class="meta">{src} â€¢ {pretty_date(dt)}</div>
                    <a class="open" href="{url}" target="_blank">Open article â†—</a>
                  </div>
                """, unsafe_allow_html=True)

st.subheader("ğŸ’¬ Ask a Question")
user_q = st.text_input("Your question about the news:")
if user_q:
    with st.spinner("Thinking..."):
        resp = graph.invoke({"query": user_q, "k": k, "mode": "", "results": [], "answer": ""})
    st.markdown("### ğŸ§  Answer")
    st.write(resp["answer"])
```
